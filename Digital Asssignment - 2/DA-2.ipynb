## Tokenizing from Web Page

```python
##Information of a complete web page
#bs4 module to crawl webpage
from bs4 import BeautifulSoup
import urllib.request
#requesting php.net for information

response = urllib.request.urlopen('https://en.wikipedia.org/wiki/India')
html = response.read()
#cleaning grabbed text
soup = BeautifulSoup(html,"html5lib")
text = soup.get_text(strip=True)
tokens_web = word_tokenize(text)
print("Tokens for this web page are: ",tokens_web[:])
#declare a dictionary
word_freq = {}
for tok in tokens_web:
    tok = tok.split()
    for t in tok:
        if t in word_freq:
            word_freq[t] += 1
        else:
            word_freq[t] = 1
```

## 3. Write a program to tokenize a sentence and find the part of speech of that sentence.

```python
nltk.download('averaged_perceptron_tagger')
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
stop_words = set(stopwords.words('english'))

string_sentence = str(input("Enter a sentence: "))

tokenized = sent_tokenize(string_sentence)
for i in tokenized:
    wordsList = nltk.word_tokenize(i)  ## these are used for finding the words and punctuation in string
    wordsList = [w for w in wordsList if not w in stop_words] ## removing the stop words
    
    tagged = nltk.pos_tag(wordsList) ## for tagging parts of speech
    print("\n")
    print(tagged)
```

```python
import string
def clean_text(text):
    # lower text
    text = text.lower()
    # tokenize text and remove puncutation
    text = [word.strip(string.punctuation) for word in text.split(" ")]
    # join all
    text = " ".join(text)
    return(text)

a=input("Enter your paragraph")

l1 = clean_text(a)
#prints para after removing punctuation
print("paragraph after removing the puctuation: ",l1)
```

```python
from nltk.corpus import stopwords
def clean_text1(l1):
    # tokenize text and remove puncutation
    l1 = [word.strip(string.punctuation) for word in l1.split(" ")]
    # remove stop words
    stop = stopwords.words('english')
    l1 = [x for x in l1 if x not in stop]
    # join all
    l1 = " ".join(l1)
    return(l1)

l2 = clean_text1(l1)
#prints para after removing stopwards
print("\nPragraph after removing the stopwords: ",l2)
```
